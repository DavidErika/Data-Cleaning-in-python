```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
DataFrame = pd.read_csv('banking_dirty.csv')
print(DataFrame)
DataFrame.dtypes
```

```python
#Deleting the 'Unnamed' column
Data = DataFrame.drop(DataFrame.columns[0], axis=1)
Data
```

```python
#Check for duplicates in Cust_id Column
DataFrame.duplicated(subset='cust_id', keep=False)
Data_clean = Data.drop_duplicates(subset='cust_id')
Data_clean
```

```python
#Check for Null values in Data_clean DataFrame
Null = Data_clean.isnull().sum()
Null
```

```python
#Change data type of birth_date, account_opened, last_transaction columns from object to datetime
Data_clean['birth_date'] = pd.to_datetime(Data_clean['birth_date'], format='%d-%m-%Y')
Data_clean['account_opened'] = pd.to_datetime(Data_clean['account_opened'], format='%d-%m-%Y')
Data_clean['last_transaction'] = pd.to_datetime(Data_clean['last_transaction'], format='%d-%m-%Y')
Data_clean['inv_amount'] = Data_clean['inv_amount'].astype('float')
Data_clean
```

```python
#Look for minimun and maximun, mode ages in the dataset
#Round all float columns to 2-decimal places
min = Data_clean['Age'].min()
max = Data_clean['Age'].max()
mode = Data_clean['Age'].mode()
mean = Data_clean['Age'].mean()
median = Data_clean['Age'].median()
print(f'minimun Age is: {min}')
print(f'maximun Age is: {max}')
print(f'mode Age is: {mode}')
print(f'Average age is: {mean}')
print(f'modian Age is: {median}')

Data_clean['fund_A'] = Data_clean['fund_A'].round(2)
Data_clean['fund_B'] = Data_clean['fund_B'].round(2)
Data_clean['fund_C'] = Data_clean['fund_C'].round(2)
Data_clean['fund_D'] = Data_clean['fund_D'].round(2)
Data_clean['acct_amount'] = Data_clean['acct_amount'].round(2)
Data_clean['inv_amount'] = Data_clean['inv_amount'].round(2)
Data_clean
```

```python
#Sort the dataframe using cust_id column
Data_clean = Data_clean.sort_values(by='cust_id')
Data_clean
```

```python
#Write cleaned data into a csv
Data_clean.to_csv('cleaned_bankingdata.csv', index=False)
```

```python
#histogram to visualise the distribution of ages in the dataset
plt.figure(figsize=(8,4))
plt.title('Age distribution of Customers')
sns.histplot(Data_clean['Age'])
```

- The majority of customers are between the ages of 45 to 49. 

```python
#Box plot to check the skewness, and outliers of Age column
plt.figure(figsize=(8,4))
plt.title('Skewness of age column')
sns.boxplot(data=Data_clean, x='Age')
```

- Outlier is anything outside this range [Q1 - 1.5 IQR, Q3 + 1.5IQR]
Where: Q1 - Lower Quartile - 1/4.
       Q3 - Upper Quartile - 3/4.
       IQR - Interquartile Range - Q3 - Q1

```python
Data_clean['Age'].sort_values()
Q1 = Data_clean['Age'].quantile(0.25)
Q2 = Data_clean['Age'].quantile(0.50)
Q3 = Data_clean['Age'].quantile(0.75)
IQR = Q3-Q1

Outlier = [Q1-(1.5*IQR), Q3+(1.5*IQR)]
Outlier
```
